# Matplotlib

- Pythonでグラフやチャートを描画するライブラリ
- ヒストグラム、散布図（Scatter Plot）、箱ひげ図（Box Plot）などが表示できる
- `import matplotlib.pyplot as plt` のようにimportする

## ヒストグラム Histogram

- データの分布を表現するグラフ
- 横軸に階級（かいきゅう：データの範囲のこと）、横軸に度数（どすう：データの出現回数）をとる
- データのばらつき、最頻値（さいひんち）、外れ値などを把握できる

### 参考：確率分布 - 一様分布 - Uniform Distibution

- すべての値が等しい確率で出現する確率分布
- NumPyでは、np.random.rand 関数で、0～1の一様分布にしたがう乱数を生成できる
- np.random.uniform 関数を使えば、範囲を指定した一様分布の乱数を生成できる
- 一様分布も2つパラメータをとる
  - 下限値（かげんち、一番小さい値のこと）
  - 上限値（じょうげんち、一番大きい値のこと）

```python
import matplotlib.pyplot as plt

count = 10000
# x = np.random.rand(count) # 一様分布：Uniform Distibution
x = np.random.uniform(low=0, high=100, size=count) # Pyhtonの"名前付き引数"を使っている
# x = np.random.randn(count) # 正規分布：Normal Distibution
plt.hist(x)
plt.show()
```

### 参考：確率分布 - 正規分布 Normal Distribution

- 平均値を中心に、左右対称なベル型の確率分布
- 自然界や、社会現象に最もよく現れる分布
- 正規分布は2つのパラメータによって、分布の形状が決定する
  - 2つのパラメータ
    - 平均
    - 分散（あるいは標準偏差）


```python
import matplotlib.pyplot as plt

count = 10000
# x = np.random.randn(count) # 2つのパラメータがない、ですよね
# 平均：70, 標準偏差：5 の正規分布にしたがう乱数を作りたい
x = np.random.normal(loc=70, scale=5, size=count)
plt.hist(x)
plt.show()
```

## 箱ひげ図 Box Plot

- データの分布を5つの"要約統計量"で表現する図
  - 最大値 100%点のこと
  - 第3四分位数（だいさん、しぶんいすう） 75%点のこと
  - 中央値 50%点のこと
  - 第1四分位数（だいいち、しぶんいすう） 25%点のこと
  - 最小値 0%点のこと
- 箱ひげ図には、外れ値（outlier）が出現する
  - 第1四分位数から - 1.5 IQR、第3四分位数から - 1.5 IQR の範囲を外れたデータのこと
- IQR： Interquantile Range
  - 第1四分位数から第3四分位数までの範囲のこと
  - つまり、箱ひげ図の"箱の長さ"のこと

```python
count = 10000
# x ：一様分布（0～100）
x = np.random.uniform(low=0, high=100, size=count)

# x ：平均：50, 標準偏差：10 の正規分布に従う乱数にしたいです
y = np.random.normal(loc=50, scale=10, size=count)

# plt.hist(x) # ヒストグラム ：データの分布（どんなデータが多いのか、データの形状のようなもの）を確認できる
plt.boxplot([x, y])
plt.show()
```

### 参考：ヒストグラム vs 箱ひげ図

- 2つのグラフ（チャート）に共通するのは、1変数を扱うところ
- ヒストグラムのよいところ
  - データの分布（形状：けいじょう、形のような言葉）を直感的に理解できる
- 箱ひげ図のよいところ
  - 省スペースで描ける
  - 並べて比較しやすい
  - 5つの要約統計量を瞬時に読み取ることができる


## 散布図 - Scatter Plot

- 2変数間の関係性を表現するグラフ ※1変数ではない点に注意
- 回帰分析や相関分析の前提確認に使用する
- 相関関係、クラスター、外れ値などを発見できる


```python
count = 100
x = np.random.rand(count) # rand関数は、0～1の範囲の一様分布です。範囲は固定です。
y = np.random.rand(count)
plt.scatter(x, y)
plt.show()

# 相関係数 :  0.05407515 ※無相関
np.corrcoef(x, y)
# array([[1.        , 0.05407515],
#        [0.05407515, 1.        ]])
```

## 参考：相関（そうかん）について

- 2つのデータの関係性のこと
- 相関を示す値に、相関係数がある

### 相関係数 - Correlation Coefficient

- 2つの変数間の直線的な関係の強さ、方向を示す値
- 相関係数は、-1 ～ +1 の範囲の値をとる
  - +1 に近いほど「正の相関が強い」
    - x が増えれば、yも増える
  - -1 に近いほど「負の相関が強い」
      - x が増えれば、yは減る
  - 0 付近は「無相関」

#### 参考の参考：相関係数の計算式

$$
\rho = \frac{S_{xy}}{S_x S_y}
$$

- Sxy: XとYの共分散
- Sx: Xの標準偏差
- Sy: Yの標準偏差

```python
# 相関のある散布図
count = 100

# 数学（math）のテスト結果、化学（science）のテスト結果のサンプルデータを作ります
math = np.random.normal(loc=50, scale=5, size=count)
science = math * 1.2 + np.random.normal(loc=0, scale=4, size=count)

plt.scatter(math, science)
plt.show()

# mathとscienceの相関係数
np.corrcoef(math, science)
```

## 復習 - 線形回帰アルゴリズム

$$
\hat{y} = a x + b
$$

- 係数：a と 切片：b をデータから計算する

```python
count = 10000
math = np.random.normal(loc=50, scale=5, size=count)
science = math * 1.2 + np.random.normal(loc=0, scale=4, size=count)
plt.scatter(math, science)

# 線形回帰アルゴリズム LinearRegression
from sklearn.linear_model import LinearRegression

model = LinearRegression()

# なぜエラーになるの？
model.fit(math.reshape(-1, 1), science) # math と scienceで学習する

# 学習済みのmodelから、係数a と 切片bを調べます

print("Coef:", model.coef_)
print("Intercept:", model.intercept_) # y = 1.20831887 x -0.4879451201226246

# 回帰直線を描画する plt.plotを使う

x = np.array([math.min(), math.max()])

# モデルの係数と切片を使って計算します
# y = model.coef_ * x + model.intercept_
# predictメソッドはとても重要です。predict: 予測する（推論する）
y = model.predict(x.reshape(-1, 1))

plt.plot(x, y, color="red")
plt.show()
```

### 参考：scikit-learnの使い方

- モデルインスタンスを生成する
  - model = LinearRegression()
- モデルを学習する
  - model.fit(x, y)
- 学習済みモデルを使って、予測する（推論する）
  - y2 = model.predict(x2)

- ポイント
  - 学習済みのモデルから、"学習によって計算されたパラメータ"を取得できる
    - model.coef_
    - model.intercept_
  - "学習によって計算されたパラメータ"には、プロパティ名に _ がつく



## 予習：教師あり学習 - 分類問題（ぶんるいもんだい）



```python
# サンプルデータ：iris ※花のデータセット
from sklearn.datasets import load_iris

iris = load_iris()
iris.data.shape # (150, 4) 150行ｘ4列

iris.data # 学習データ
iris.target # 150個の教師データ 0:setosa 1:versicolor 2:virginica

# iris.target_names # 花の種類の名前 ['setosa', 'versicolor', 'virginica']

# 機械学習アルゴリズム：近傍法（kNN法）
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier()

# 150件のデータを学習
model.fit(iris.data, iris.target)

# 学習済みモデルで推論する
# 0 or 1 or 2 の結果が返る
model.predict([[2, 3, 2, 2]])
```
